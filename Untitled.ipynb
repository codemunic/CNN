```python
#Data Preprocessing
train_X=train_set_x_orig/255
test_X=test_set_x_orig/255
train_y=np.eye(6)[train_set_y_orig.reshape(-1)]
test_y=np.eye(6)[test_set_y_orig.reshape(-1)]
```

```python
# Create Placeholders
X=tf.placeholder(shape=[None, 64,64,3],dtype=tf.float32)
y=tf.placeholder(shape=[None,6],dtype=tf.float32)
```

```python
#initialize parameters
W1=tf.get_variable("W1",[4,4,3,8],initializer=tf.contrib.layers.xavier_initializer(seed = 0))
W2=tf.get_variable("W2",[2,2,8,16],initializer=tf.contrib.layers.xavier_initializer(seed = 0))
```

```python
#create model
Z1=tf.nn.conv2d(X,W1,strides=[1,1,1,1],padding='SAME') # strides=[1,s,s,1]
A1=tf.nn.relu(Z1)
P1=tf.nn.max_pool(A1,ksize=[1,8,8,1],strides=[1,8,8,1], padding='SAME')
Z2=tf.nn.conv2d(P1,W2,strides=[1,1,1,1],padding='SAME')
A2=tf.nn.relu(Z2)
P2 = tf.nn.max_pool(A2, ksize = [1,4,4,1], strides = [1,4,4,1], padding = 'SAME')
P2=tf.contrib.layers.flatten(P2)
Z3=tf.contrib.layers.fully_connected(P2,6,activation_fn=None)
```

```python
#compute cost
cost=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=Z3, labels=y))
```

```python
def random_mini_batches(X,y, minibatch_size, seed):
    m=X.shape[0]
    num_complete_minibatch=int(m/minibatch_size)
    np.random.seed(seed)
    mini_batches=[]
    permutation=list(np.random.permutation(m))
    Shuffled_X=X[permutation,:,:,:]
    Shuffled_y=y[permutation,:]
    for k in range(num_complete_minibatch):
        minibatch_X=Shuffled_X[k*minibatch_size:k*minibatch_size+minibatch_size,:,:,:]
        minibatch_y=Shuffled_y[k*minibatch_size:k*minibatch_size+minibatch_size,:]
        minibatch=(minibatch_X,minibatch_y)
        mini_batches.append(minibatch)
    if m%minibatch_size!=0:
        minibatch_X=Shuffled_X[num_complete_minibatch*minibatch_size:m,:,:,:]
        minibatch_y=Shuffled_y[num_complete_minibatch*minibatch_size:m,:]
        minibatch=(minibatch_X,minibatch_y)
        mini_batches.append(minibatch)
    return mini_batches
```

```python
m=1080
tf.set_random_seed(2)
costs=[]
optimizer = tf.train.AdamOptimizer(learning_rate=0.009).minimize(cost)
init = tf.global_variables_initializer()
minibatch_size=64
with tf.Session() as sess:
    sess.run(init)
    for epoch in range(50):
        minibatch_cost=0
        num_minibatches = int(m / minibatch_size)
        seed=0
        minibatches=random_mini_batches(train_X,train_y, minibatch_size, seed)
        for item in minibatches:
            (minibatch_X, minibatch_y) = item
            _ , temp_cost = sess.run([optimizer, cost], feed_dict={X:minibatch_X, y:minibatch_y})
            minibatch_cost+=temp_cost/num_minibatches
        costs.append(minibatch_cost)
        if epoch%5==0:
            print('Cost after epoch '+str(epoch)+' is '+ str(costs[epoch]))
            
    plt.plot(costs)
    predict_op = tf.argmax(Z3, 1)
    correct_prediction = tf.equal(predict_op, tf.argmax(y, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print(accuracy)
    train_accuracy = accuracy.eval({X: train_X, y: train_y})
    test_accuracy = accuracy.eval({X: test_X, y: test_y})
    print("Train Accuracy:", train_accuracy)
    print("Test Accuracy:", test_accuracy)
```
